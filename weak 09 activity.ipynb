# Finding out whether the same amount was spent for the three advertisements

### H0=mean amount spent for all the 3 adds are equal

### H1=mean amount spent for the 3 adds differs

```python
from scipy.stats import f_oneway
f_oneway(data['TV'],data['Radio'],data['Newspaper'])
```

Here p value almost equal to zero and less than 0.5.So that we can reject the null hypothesis.From this testing we can conclude that amount spent for 3 adds are different.
